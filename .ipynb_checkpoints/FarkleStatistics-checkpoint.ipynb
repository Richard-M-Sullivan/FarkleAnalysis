# Calculating Farkle Odds For The Purpose Of Defining Optimal Strategy #

### Preface ###

Farkle is a simple party game where players take turn rollig dice to accumulate points. That on it's own would make for a boring game, but the rules allow the ability to stop rolling and keep the points accumulated while simultaneously passing the remaining dice to the next player. This is important because each dice roll must meet a set of conditions or you lose all of your points for that round. To prevent this you have to pass, which allows you to keep your points but ends your turn. Subsiquently, the next player then has the option to roll the remaining dice passed (which may be few, and thus giving smaller odds to meet the conditions) in order to add the previous players score to their own, or they can chose to start over and roll all the dice from scratch.

### Finding The Odds Of Passing A Roll ###

In the game of Farkle there is a maximum of six dice. Because the number of possible combinations is so small, the function will iterate through every combination while applying the Farkle rules to determine if the roll is a success or a failure. For each successful roll, a counter will be updated to record winning combinations. Afterwards, the function will divide that number by the total number of combinations to produce the odds of a successful role. 

```python
def passingRolePercent(numDice):
    
    dice = [1 for i in range(numDice)] #create a list of dice that are initialized to be all ones
    totalCombinations = 6 ** numDice  #record the total number of combinations
    
    wins = 0 # number of wins
    
    while (dice[numDice-1] != 7):# check if the dice combination passed all 6's
        
        # check if the combination is a win or not
        
            #catch 1 or 5 rolled or a run of 1 - 6
        if(1 in dice or 5 in dice):
            wins += 1
        else:
            #catch -> 3, 4, 5, 6 of a kind or 4 of a kind and pair or two triplets
            if len(dice) >= 3:
                for i in range(1,7):
                    if dice.count(i) > 2:
                        wins += 1
                        break
                #catch -> three pairs
                if (len(dice) == 6):
                    list = [dice.count(i) for i in range(1,7)]
                    if (list.count(2) == 3):
                        wins += 1
                
                
        
        #incrament the first die
        dice[0] += 1
        
        # propogate through the rest of the dice
        for i in range(numDice):
            if dice[numDice-1] == 7:
                break
            elif dice[i] == 7: #if a die has number too high, then reset it to 1 and increase the next one by 1
                dice[i] = 1
                dice[(i+1) % numDice] += 1
                
    return wins / totalCombinations
```

```python
#creating list of x and y variables to capture input and output of the passing role percent funciton
x = []
y = []

#generating and printing out the values 
for numOfDice in range(TOTAL_NUM_OF_DICE):
    x.append(numOfDice+1)
    y.append(passingRolePercent(x[numOfDice])*100)
    print("odds of not losing with %d dice: %.2f%%" % (numOfDice+1,y[numOfDice]))
    
```

As one would expect, the odds of having a successful role go down as you have fewer dice. The odds plotted onto a chart looks like the following â€“ to help visualize the relationship between dice and your chances of success.

```python
plt.bar(x,y)
plt.title("Win Percentage vs Dice")
plt.xlabel("Number of Dice")
plt.ylabel("Chance of Winning")
plt.show()
```

### Finding The Average Attainable Score ###

When playing Farkle it is often neccessary to decide whether to roll the remaining dice of the previous player for a chance to build off of their score or to start from scratch. Of course, if the previous player passes a low score with few remaining die then it is assumed that it would not be advantageous to build off of their score, for the risk would be too high for the low reward. If one is passed a large score with many dice the chance of winning those points would be more favorable than in the previous scenario. In order to assess where the line might be for deciding whether to take or pass on the previous players score, one would need to know the likelyhood of success to win the role as well as understanding how many points one could make if starting from scratch. 

For finding the average score, when starting from scratch, a simulation will be used. In the simulation, the player will try to achieve the maximum score theoretically possible. This sounds simple in theory, but there are multuple ways one could go about doing this. By having the ability to choose what dice to keep and what dice to re-roll, there is room for several strategies. One strategy would be to roll the dice, and keep all of dice that award points (this will be referred to as the '*greedy*' strategy). Another method would be to take only minimum number of dice that award the maximum number of points (This will be referred to this as the '*timid*' strategy) The last method that will be tried out will be called the '*hybrid*' strategy, in which the player takes the dice with the points that would allow for the highest odds for a large score. This differs from the timid method in which it might play timid while having a large number of dice in an attempt to get a high scoring combination, but when those combinations are no longer possible it may play greedy to take the points of the remaining number of dice. 

By employing each strategy and running simulations of their performance it may be reveiled what strategy yeilds the highest average score.

#### Greedy Strategy ####

The greedy function takes in a number of dice. Then the greedy function will roll the dice and calculate the total number of points available. Afterwards it will roll the remaining dice and keep adding to its score until it loses. Lastly the greedy function will pass back its final score.

```python
def greedy(diceNum):
    dice =[random.randint(1,6) for die in range(diceNum)] #roll and create dice
    dice.sort() #sort dice
    
    frequency = [0,0,0,0,0,0] #store how many times the possible numbers come up
    for die in dice:
        frequency[die -1] += 1
    
    score = 0 # create a container to hold the score
    
    if len(dice) == 6: # if all 6 dice look for big scoring combinations
        if frequency.count(6) == 1 : #six of a kind
            score += 3000
            dice = []
        elif frequency.count(3) == 2 :#two triples
            score += 2500
            dice = []
        elif frequency.count(2) == 3:#three pairs
            score += 1500
            dice = []
        elif frequency.count(4) == 1 and frequency.count(2) == 1:#four of a kind and a pair
            score += 1500
            dice = []
        elif frequency.count(1) == 6: # strait
            score += 1500
            dice = []
    
    if len(dice) > 4: # if there are more than 4 dice
        if frequency.count(5): #5 of a kind
            dieNum = frequency.index(5)
            frequency[dieNum] = 0
            for i in range(5):
                dice.remove(dieNum+1)
            score += 2000
            
    if len(dice) > 3: #if there are more than 3 dice
        if frequency.count(4): # four of a kind
            dieNum = frequency.index(4)
            frequency[dieNum] = 0
            for i in range(4):
                dice.remove(dieNum+1)
            score += 1000
            
            
    if len(dice) > 2 : #if there are three or more dice scan for having 3 of a kind
        if frequency.count(3): #three of a kind
            dieNum = frequency.index(3)
            frequency[dieNum] = 0
            for i in range(3):
                dice.remove(dieNum+1)
            score += (dieNum+1) * 100
            
    
    if len(dice) > 0: # with the remaining dice look for ones and fives
        if dice.count(1) > 0:
            score += dice.count(1) * 100
            for i in range(dice.count(1)):
                dice.remove(1)
                
        if dice.count(5) > 0:
            score += dice.count(5) * 50
            for i in range(dice.count(5)):
                dice.remove(5)

    return {'score':score,'remaining':dice}
```

#### Timid Strategy ####

The timid fuction takes in a number of dice. Then the timid function will roll the dice and calculate the largest number of points possible with the minimum number of die. The way this will work is by dviding the points earned by the number of die to get a point potency. It uses this information to make the choice with the highest potency whilst using the minimum number of die. This ensures that if there are multuple options with the same point potency, that it will take the option which uses up the lowest number of dice. It will continue playing in this manner until it loses. Lastly the timid function will pass back its final score.

```python
def timid(diceNum):
    initDiceNum = diceNum
    dice =[random.randint(1,6) for die in range(diceNum)] #roll and create dice
    dice.sort() #sort dice

    bigCombination = False
    bigThree = False
    take3 = False
    takeOne = False
    
    frequency = [0,0,0,0,0,0] #store how many times the possible numbers come up
    for die in dice:
        frequency[die -1] += 1
    
    score = 0 # create a container to hold the score
    
    if len(dice) == 6: # if all 6 dice look for big scoring combinations
        if frequency.count(6) == 1 : #six of a kind
            score += 3000
            dice = []
            bigCombination = True
        elif frequency.count(3) == 2 :#two triples
            score += 2500
            dice = []
            bigCombination = True
        elif frequency.count(2) == 3:#three pairs
            score += 1500
            dice = []
            bigCombination = True
        elif frequency.count(4) == 1 and frequency.count(2) == 1:#four of a kind and a pair
            score += 1500
            dice = []
            bigCombination = True
        elif frequency.count(1) == 6: # strait
            score += 1500
            dice = []
            bigCombination = True
    
    if len(dice) > 4: # if there are more than 4 dice
        if frequency.count(5): #5 of a kind
            dieNum = frequency.index(5)
            frequency[dieNum] = 0
            for i in range(5):
                dice.remove(dieNum+1)
            score += 2000
            bigCombination = True
            
    if len(dice) > 3: #if there are more than 3 dice
        if frequency.count(4): # four of a kind
            dieNum = frequency.index(4)
            frequency[dieNum] = 0
            for i in range(4):
                dice.remove(dieNum+1)
            score += 1000
            bigCombination = True
            
    
            
    if bigCombination == False:       
        if len(dice) > 2 : #if there are three or more dice scan for having 3 of a kind
        
            if frequency.count(3): #three of a kind
                dieNum = frequency.index(3) 
            
                # if we find a three of a kind we need to know if it worth taking
                if dieNum +1 >= 4: #if it is better than just getting ones take the combination of three
                    big3 = True
                    take3 = True
                elif len(dice) == frequency[0]+frequency[4]+3: # you will be able to clear the board then take it
                    take3 = True
                elif (len(dice) > 3) and (frequency[0] + frequency[4] > 0): #if you have the option to take one die
                    pass                                           # with dice remaining to make another triple
                else:  # if the triple is your only option then take it
                    take3 = True
            
                if take3 == True: # if it is ok to take the triple then take it
                    frequency[dieNum] = 0
                    for i in range(3):
                        dice.remove(dieNum+1)
                    score += (dieNum+1) * 100
            
        #if you can clear the rest of the dice 
        if len(dice) == frequency[0]+frequency[4]:
            score += frequency[0] * 100
            for i in range(frequency[0]):
                dice.remove(1)
            score += frequency[4] * 50
            for i in range(frequency[4]):
                dice.remove(5)
        elif take3 == True and len(dice) < 3:
            if frequency[0] > 0:
                score += frequency[0] * 100
                for i in range(frequency[0]):
                    dice.remove(1)
        elif len(dice) <= 3 and take3 == False: #no way to make triple next round
            if frequency[0] > 0:
                score += frequency[0] * 100
                for i in range(frequency[0]):
                    dice.remove(1)
            else:
                if frequency[4] > 0:
                    score += 50
                    dice.remove(5)
                    
        elif take3 == False: # if you didn't take three
            if len(dice) > 0: 
                if frequency[0] > 0:
                    score += 100
                    dice.remove(1)
                
                elif frequency[4] > 0:
                    score += 50
                    dice.remove(5)
    else:
        if len(dice) == frequency[0]+frequency[4]: #if you can clear the rest of the dice
            score += frequency[0] * 100
            for i in range(frequency[0]):
                dice.remove(1)
            frequency[0] = 0
            score += frequency[4] * 50
            for i in range(frequency[4]):
                dice.remove(5)
            frequency[4] = 0

    return {'score':score,'remaining':dice}
```

##### Hybrid Strategy ####

The hybrid function takes in a number of dice. Then the hybrid function will employ the greedy or timid strategy based on the number of dice available. If the number of dice is below a set pivot then the greedy strategy will be used, or in the other case it will use the timid method.

```python
def hybrid(diceNum,pivot):
    piv = pivot
    
    if diceNum < piv:
        return greedy(diceNum)
    else:
        return timid(diceNum)
```

###### Comparing The Greedy and Timid Strategy ######
For hybrid function to work, it needs a metric to make decisions on how it should go about playing. The solution for this would be to compare the greedy strategy with the timid strategy to see in what scenerios one would be preferred to the other. Based on this information one could merge the two to make a final strategy that is better than the sum of it's parts.

```python
greedyAverage =[0,0,0,0,0,0]
timidAverage = [0,0,0,0,0,0]

x = [6,5,4,3,2,1]
for trial in range(len(greedyAverage)):
    for i in range(100000+1):
        dice = 6 - trial
        while dice > 0:
            info = greedy(dice)
            dice = len(info['remaining'])
            greedyAverage[trial] += info['score']
    greedyAverage[trial] = greedyAverage[trial] / 100000

    
for trial in range(len(timidAverage)):
    for i in range(100000+1):
        dice = 6 - trial
        while dice > 0:
            info = timid(dice)
            dice = len(info['remaining'])
            timidAverage[trial] += info['score']
    timidAverage[trial] = timidAverage[trial] / 100000

print(greedyAverage)
print(timidAverage)
```

```python
plt.plot(x,greedyAverage,label="greedy strategy")
plt.plot(x,timidAverage,label="timid strategy")
plt.title("Average Score VS Starting Number of Dice")
plt.xlabel("Number of Dice at Start")
plt.ylabel("Average Score")
plt.legend()
plt.show()

table = [[x for x in reversed(greedyAverage)],[x for x in reversed(timidAverage)]]
pd.DataFrame(table,index=['greedy point average', 'timid point average'],columns=['1','2','3','4','5','6'])
```

When looking at the plots above, it is observed that the timid strategy dominates when the number of dice is high, but barely underperforms to the greedy strategy when the number of dice is less than three. As the number of dice goes down the timid funciton converges with the greedy function. In order to understand why this is the case one must learn about the potential of getting points at different dice numbers.

##### Exploring the Point Potency of a Single Roll With Varying Numbers of Dice #####

The singleRollPoints function rolls the dice set at a specific number and calculates the maximum score available. 

```python
#returns the maximum number of points available in a simulated dice roll
def singleRollPoints(diceNum):
    return greedy(diceNum)['score']
```

```python

y = 100000
x = [ [singleRollPoints(i) for j in range(y+1)] for i in range(1,7)]

bins = [i*50 for i in range(3000//50)]
fig, ax = plt.subplots(nrows=3,ncols=2)

for i in range(6):
    ax[i//2][i%2].set_title('Probability of Single Roll Score With '+str(i+1)+" Dice")
    ax[i//2][i%2].set_xlabel('points')
    ax[i//2][i%2].set_ylabel('occurences out of '+str(y)+' rolls')
    ax[i//2][i%2].hist(x[i],bins)
    ax[i//2][i%2].axvline(sum(x[i])/y, color='#fc3f30', label='median score\n'+str(sum(x[i])/y),linewidth=2)
    ax[i//2][i%2].legend()

plt.tight_layout()
plt.subplots_adjust(left=0.5,right=2,bottom=0.2,top=2.9,wspace=.25,hspace=.5)
plt.show()
```

As we increase the number of dice, the potential score increases by a considerable ammount â€“ almost doubling each time. This shows that one is most likely to get the most points is with four or more dice. This is because for these numbers of dice grant access to the largest number of high scoring combinations. It is also important to note that having all of the dice is the best by a signifigant margin. Being able to reach that point will almost assure that you will have a large score no matter what number of dice you start off with. This explains the differences found in the timid and greedy strategies. The timid function lingers and allows for the most attempts at a high scoring combination that requires higher numbers of dice, which maximizes the score possible as you go down to one die. The greedy function works by trying to get back to rolling all six dice as many times as possible. When the timid function starts with lots of dice the score is larger than the greedy strategy, for both strategies are unlikely to go back to having six again and the timid function is more efficient at getting larger point values. When the number of dice is low, then the number of points to be made is also low. This explains why the greedy function wins because it is more capable of getting back to having all six dice (a chance of getting a large score) than the timid dice are. But we see that the scores are very close when you get down to three or fewer dice. The greedy function when it is able to get to six dice is not able to capitalize on having that large ammount as well as the timid strategy, but despite this handycap it still does well enough to surpass the timid function within this range. This suggests that the hybrid function will be able to out perform both strategies by a considerable and noticable margin by using the best parts of both strategies.

##### Finding The Pivot #####

When comparing the greedy and timid strategies we saw that they became very close at three dice and the greedy function slightly surpassed the timid one when reaching less than three dice. Keeping this in mind a good pivot value would be four. This means that one should use the timid strategy, but when the number of dice is less than four one should switch to the greedy strategy. For the purposes of finding the optimal pivot, all pivot values within the range 2-6 will tested for performance and then compared to select the best one. 

```python
pivotAverage = [0,0,0,0,0]
x =            [2,3,4,5,6]

hybridAverage = [0,0,0,0,0,0]

trials = 1000000

for piv in range(len(x)):
    for i in range(6):
        for trial in range(trials):
            dice = i+1
            while dice != 0:
                info = hybrid(dice,piv+2)
                hybridAverage[i] += info['score']
                dice = len(info['remaining'])
        hybridAverage[i] = hybridAverage[i]/trials
    pivotAverage[piv] = hybridAverage.copy()


for table in pivotAverage:
    data = ""
    for item in table:
        print("%.2f" % item,end=" ")
    print()
    
```

In the graph the timid strategy does not show up, but this is due to the hybrid strategy being so close that they are approximately the same value. The pivot of 2 was chosen because it yeilded the highest results, while being different from the timid strategy, and it nets in an average that is just slightly higher than the timid strategy. Whether or not this is useful in terms of strategy is hard to know because all of the calculations are for the theoretical maximum point average that one could hope to achieve.
